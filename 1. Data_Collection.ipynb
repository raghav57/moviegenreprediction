```python
def youtubeExtract(id):
    from youtube_transcript_api import YouTubeTranscriptApi
    try:
        transcript = YouTubeTranscriptApi.get_transcript(id)
        trailerTranscript = ' '.join([transcript[i]['text'] for i in range(len(transcript))])
        return trailerTranscript
    except:
        return ''

youtubeId['trailerTranscript'] = youtubeId.apply(lambda row:youtubeExtract(row['youtubeId']),axis=1)
```
